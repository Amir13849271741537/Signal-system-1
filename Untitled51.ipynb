```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
```

```python

def generate_damped_sinusoid(
    M=512,
    A=1.0,
    alpha=1.0,
    nu=3.3,
    phi=0.0,
    SNR_dB=40
):
    """
    Generate real-valued damped sinusoid + white Gaussian noise
    based on equation (1) of the paper.
    """

    m = np.arange(M)

    signal = A * np.exp(-(2*np.pi/M)*alpha*m) * \
             np.cos((2*np.pi*nu/M)*m + phi)

    signal_power = A**2 / 2
    noise_power = signal_power / (10**(SNR_dB/10))

    noise = np.random.normal(
        0,
        np.sqrt(noise_power),
        M
    )

    y = signal + noise

    return m, signal, noise, y
```

```python

M = 512
A = 1
alpha = 1
nu = 3.3
phi = np.pi / 4
SNR = 40

m, clean_signal, noise, y = generate_damped_sinusoid(
    M, A, alpha, nu, phi, SNR
)
```

```python

plt.figure(figsize=(12,4))
plt.plot(m, y, label='Noisy Signal')
plt.plot(m, clean_signal, label='Clean Signal', linewidth=2)
plt.legend()
plt.grid()
plt.title("Damped Sinusoid")
plt.show()
```

```python

def hann_window(M):
    n = np.arange(M)
    return 0.5 - 0.5*np.cos(2*np.pi*n/(M-1))
```

```python

window = hann_window(M)
yw = y * window
```

```python

Y = fft(yw)
magnitude = np.abs(Y[:M//2])
```

```python

plt.figure(figsize=(12,4))
plt.plot(magnitude)
plt.grid()
plt.title("Spectrum Magnitude")
plt.xlabel("Bin")
plt.ylabel("|Y(k)|")
plt.show()
```

```python

l = np.argmax(magnitude)
print("Peak bin l =", l)
```

```python
# 🟢 Part 2 — IpDFT Algorithm Implementation
```

```python

def estimate_s(Y, l):
    left = np.abs(Y[l - 1])
    right = np.abs(Y[l + 1])

    if right > left:
        return 1
    else:
        return -1
```

```python

def compute_rho(Y, l, s):
    if s == 1:
        rho = Y[l + 1] / Y[l]
    else:
        rho = Y[l - 1] / Y[l]

    return rho
```

```python

def ipdft_estimator(rho, s, H=2):
    estimate = s * ((H * rho) + (H - 1)) / (rho - 1)

    delta_hat = np.real(estimate)
    alpha_hat = np.imag(estimate)

    return delta_hat, alpha_hat
```

```python

def ipdft(Y, l, H=2):
    s = estimate_s(Y, l)

    rho = compute_rho(Y, l, s)

    delta_hat, alpha_hat = ipdft_estimator(
        rho,
        s,
        H
    )

    return delta_hat, alpha_hat, s, rho
```

```python

delta_hat, alpha_hat, s, rho = ipdft(Y, l)

nu_hat = l + delta_hat

print("Estimated s =", s)
print("Estimated delta =", delta_hat)
print("Estimated frequency =", nu_hat)
print("Estimated alpha =", alpha_hat)
```

```python

print("========== Ground Truth ==========")
print("True nu =", nu)
print("True alpha =", alpha)

print("\n========== Estimated ==========")
print("Estimated nu =", nu_hat)
print("Estimated alpha =", alpha_hat)

print("\n========== Errors ==========")
print("Frequency error =", abs(nu_hat - nu))
print("Alpha error =", abs(alpha_hat - alpha))
```

```python

import numpy as np
import matplotlib.pyplot as plt

# تنظیمات اولیه
fs = 1000  # فرکانس نمونه برداری
t = np.linspace(0, 1, fs, endpoint=False)

# سیگنال ترکیبی (دو فرکانس + نویز)
f1 = 50
f2 = 120

signal = np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)
signal += 0.2 * np.random.randn(len(t))  # نویز
```

```python

plt.figure()
plt.plot(t, signal)
plt.title("Signal in Time Domain")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python

fft_result = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), 1/fs)

# فقط نیمه مثبت طیف
mask = freqs >= 0
freqs = freqs[mask]
magnitude = np.abs(fft_result)[mask]
```

```python

plt.figure()
plt.plot(freqs, magnitude)
plt.title("Frequency Spectrum (FFT)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
